```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
mod_name = "huggingface/CodeBERTa-small-v1"
tokenizer = AutoTokenizer.from_pretrained(mod_name)
model = AutoModel.from_pretrained("huggingface/CodeBERTa-small-v1")
```

```python
user_input = input("Enter a natural language or code snippet: ")

# Tokenize
inputs = tokenizer(user_input, return_tensors="pt", truncation=True, padding=True)

# Generate embeddings
with torch.no_grad():
    outputs = model(**inputs)

# CLS token (first token) embedding
cls_embedding = outputs.last_hidden_state[:, 0, :]

# Mean pooling (recommended)
mean_embedding = outputs.last_hidden_state.mean(dim=1)

print("\n--- CLS Embedding ---\n", cls_embedding)
print("\n--- Mean Embedding ---\n", mean_embedding)
```
